### 1) Introduction to Lists

```python
## Create a list 
mylist = [] # the list is empty
mylist = [1, 2, 3, 4, 5] # the list has 5 elements

## Access elements in the list
print(mylist[0]) # access the first element

## Length of the list
print(len(mylist))

## Add elements to the list
mylist.append("a") #we can add different types of elements to the list
print(type(mylist[0]))
print(type(mylist[5]))

## Remove elements from the list
mylist.remove("a") # remove the first occurence of "a"
del mylist[0] # remove the first element
print(mylist)

## Slicing
print(mylist[1:3]) # get the second and third elements

## Concatenate lists
mylist.extend([6, 7, 8]) # add multiple elements to the list
print(mylist)
```

### 2) Use of lists

```python
mylist1 = ['a', 3, 12, 3, 'a', 3]

## Count the number of occurences of an element
print(mylist1.count('a')) # 2 times

## Find the index of an element
print(mylist1.index(3)) # 1

## Reverse the list
mylist1.reverse()
print(mylist1)

## Concatenate 2 lists
mylist2 = [1, 2, 'b']
mylist = mylist1 + mylist2
print(mylist)

## Insert an element at a specific index
mylist.insert(4, 'b')
print(mylist)

## Sort the list
#mylist.sort() # the list should have elements of the same type
newlist = [4,5,8,1,0,34,65,2,7]
newlist.sort()
#Or
sorted(newlist)
print(newlist)

#Descending order
newlist.sort(reverse=True)
#Or
sorted(newlist, reverse=True)

## Create a list from a list
newlist1 = newlist[1:6]
print(newlist1)

## Nested lists
nestedlist = [[1,2,3], [4,5,6], [7,8,9]] # a list of lists

## Access elements in a nested list
print(nestedlist[0][0]) # access the first element of the first list

## Loop through a list
for i in range(len(newlist)):
    print(newlist[i])

## Transform a list into a string
mylist = ['a', 'b', 'c']
string = ''.join(mylist)
print(string)

## Transform a string into a list
string = 'Il/court/dans/le/parc/avec/ses amis'
mylist = list(string)
#Or
mylist = string.split("/") # split the string at each occurence of "/". Automatic with space
print(mylist)

## Transform a type of an element
newlist[3] = float(newlist[3])
print(type(newlist[3]))
```

### 3) Dictionnaries

```python
## Create a dictionary
mydict = {} # the dictionary is empty
mydict = {'a': 1, 'b': 2, 'c': 3} # the dictionary has 3 key-value pairs

## Access elements in the dictionary
print(mydict['a']) # access the value of the key 'a'
#Or
print(mydict.get('b')) 

## Length of the dictionary
print(len(mydict))

## Add elements to the dictionary
mydict['d'] = 4
print(mydict)

## Remove elements from the dictionary
del mydict['a']
#Or
mydict.pop('d')
print(mydict)

## Check if an element is in the dictionary
'e' in mydict # False
'c' in mydict # True

## Access the keys of the dictionary
print(mydict.keys())

## Access the values of the dictionary
print(mydict.values())

## Access the keys and values of the dictionary
print(mydict.items())

## Loop through a dictionary
for key in mydict:
    print("La lettre",key, "est associé au chiffre",mydict[key])

## Update a dictionary
mydict.update({'b': 1, 'e': 5, 'f': 6}) # add new key-value pairs and update existing ones
print(mydict)

## Create a dictionary from 2 lists
keys = ['a', 'b', 'c']
values = [1, 2, 3]
mydict1 = dict(zip(keys, values))
print(mydict1)


```

### 4) Application of comprehensions and generators

```python
# WITH LISTS

## Loops for even numbers
list = range(20)

mylist = []
for x in list:
    if x % 2 == 0:
        mylist.append(x)
print(mylist)

## Same thing with list comprehension
mylist2 = [x for x in list if x % 2 == 0]
print(mylist2)

## Convert int to string
mylist2 = [str(x) for x in mylist2]
print(mylist2)

######################################################################################
# WITH DICTIONARIES - Comprehension


## Square of even numbers with dictionary comprehension
mydict = {x: x**2 for x in list if x % 2 == 0}
print(mydict)

## Switch keys and values
dict = {'Justine': 'Student', 'Marie': 'Teacher', 'Paul': 'Engineer'}
newdict = {value: key for key, value in dict.items()}
print(newdict)


## WITH DICTIONARIES - Generators

# Compare comprehension and generator
C = [x**2 for x in range(10)] # comprehension
G = (x**2 for x in range(10)) # generator
print(C)
print(G) # the generator is not exhausted

print(sum(C))
print(sum(G)) # the generator is exhausted
```
